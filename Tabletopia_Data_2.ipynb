```python
import pandas as pd
bgg_data = pd.read_csv("C:/Users/Rob/Downloads/board-game-data/bgg_db_2018_01.csv", encoding = "ISO-8859-1")
bgg_data.head()
```

```python
with open('C:/Users/Rob/Downloads/board-game-data/parsed_data.txt', 'r') as f:
    tt_data = f.readlines()
```

```python
tt_data = tt_data[0].split(',')
tt_data = tt_data[:-1]
tt_data
```

```python
game_ids = []
for address in tt_data:
    string = address.split('/')
    game_ids.append(int(string[4]))
game_ids
```

```python
bgg_data['game_id'].dtypes
```

```python
match = bgg_data.game_id.isin(game_ids)
match.shape
```

```python
bgg_data['tt_match'] = match
bgg_data.head()
```

```python
results = bgg_data.loc[bgg_data['tt_match'] == True]
#results = bgg_data.loc[bgg_data['names'].str.lower().isin(lower_bga)]
#results.shape
results.head()
```
